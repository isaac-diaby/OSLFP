## Getiing static image of the building via Google map API

```python
static_folder = f"./testing"
static_path = f"{static_folder}/main.png"
lat, lng = geocode_result[0]['geometry']['location'].values()
print(lat, lng)
```

```python
dirExist = os.path.exists(static_folder)
print(dirExist)
if not dirExist:
    os.makedirs(static_folder)
```

```python

with open(static_path, 'wb') as out:
    for chunk in gmaps.static_map(size=(400, 400),
                                center=(lat,lng),
                                zoom=17,
                                maptype='satellite'):
        if chunk:
            out.write(chunk)
```

## Get near by building informations form the geocode

```python
nearby = gmaps.places_nearby(location=(lat,lng), radius=100)
nearby
```

```python
m = folium.Map(location=[lat, lng], zoom_start=15, zoom_control=True)
m
```
